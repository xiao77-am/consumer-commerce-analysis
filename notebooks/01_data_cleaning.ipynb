```python
#数据清洁
import pandas as pd

badge_batch1 = pd.read_excel("../data/raw/徽章一批.xlsx")
badge_batch2 = pd.read_excel("../data/raw/徽章二批.xlsx")
baseball = pd.read_excel("../data/raw/棒球服.xlsx")
badge_extra = pd.read_excel("../data/raw/徽章余量.xlsx")
```

```python
print("徽章一批：", badge_batch1.columns.tolist())
print("徽章二批：", badge_batch2.columns.tolist())
print("棒球服：", baseball.columns.tolist())
print("徽章余量：", badge_extra.columns.tolist())
```

```python
orders = pd.concat(
    [badge_batch1, badge_batch2, baseball, badge_extra],
    ignore_index=True
)
```

```python
orders.shape
```

```python
orders.head()
```

```python
orders.to_excel("../data/processed/orders_clean.xlsx", index=False)
```

```python
# 转换发货时间为日期类型
orders["发货时间"] = pd.to_datetime(orders["发货时间"],errors="coerce")

# 转换支付时间为日期类型
orders["支付时间"] = pd.to_datetime(orders["支付时间"])

# 订单编号转换为字符串
orders["订单编号（勿删）"] = orders["订单编号（勿删）"].astype(str)
```
